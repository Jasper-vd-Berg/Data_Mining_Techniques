```python
#pandas 1.4.1
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder
odi_csv = pd.read_csv('ODI-2022.csv',sep=';')
```

```python
col_names = odi_csv.columns
#Tijdstempel date always same, time different, but time column probably not important
#What programme are you in ? many same but different values (AI, Ai, Mcs AI, Mc ai, Artificial intelligence etc.)
#Lots of features only occur once.
#Have you taken a course on machine learning (categorical yes/no) with some missing values (unknown)
#Have you taken a course on information retrieval (binomial 0/1) with some missing values (unknown)
#Have you taken a course on statistics (categorical mu/sigma) with some missing values (unknown)
#Have you taken a course on databases (categorical ja/nee) with some missing values (unknown)
#What is your gender (categorical 5 categories, no missing data, unequal class distribution) (not important)
#Chocolate makes you... (categorical 5 categories) some missing values (unknown) (not important)
#When is your birthday (many different values, many features only occur once) (not important)
#Number of neigbors sitting around you (see when is your birthday)
#long column name (not relevant)
#give a random number (not relevant)
#time you went to be yesterday (not relevant)
#what makes a good day for you (not relevant)
for col in col_names:
    print('\n',col,'\n')
    counts = odi_csv.loc[:,col].value_counts()
    names = odi_csv.loc[:,col].value_counts().index.to_list()
    for name,count in zip(names,counts):
        print(name,'\t\t',count)
```

```python
odi_cleaned = odi_csv.replace('unknown',np.NaN)
odi_cleaned.drop(['Tijdstempel', 
                  'What is your gender?',
                  'Chocolate makes you.....',
                  'When is your birthday (date)?',
                  'Number of neighbors sitting around you?',
                  'Did you stand up?',
                  'What is your stress level (0-100)?',
                  'You can get 100 euros if you win a local DM competition, or we don’t hold any competitions and I give everyone some money (not the same amount!). How much do you think you would deserve then? ',
                  'Give a random number',
                  'Time you went to be Yesterday',
                  'What makes a good day for you (1)?',
                  'What makes a good day for you (2)?'], axis=1,inplace=True)
odi_cleaned.rename(columns = {'What programme are you in?':'Current Program',
                              'Have you taken a course on machine learning?':'Machine Learning',
                              'Have you taken a course on information retrieval?':'Information Retrieval',
                              'Have you taken a course on statistics?':'Statistics',
                              'Have you taken a course on databases?':'Databases'}, inplace = True)
binomial = {'yes': 1,'no': 0,np.nan:np.nan}
odi_cleaned['Machine Learning'] = [binomial[item] for item in odi_cleaned['Machine Learning']]
binomial = {'1': 1,'0': 0,np.nan:np.nan}
odi_cleaned['Information Retrieval'] = [binomial[item] for item in odi_cleaned['Information Retrieval']]
binomial = {'mu': 1,'sigma': 0,np.nan:np.nan}
odi_cleaned['Statistics'] = [binomial[item] for item in odi_cleaned['Statistics']]
binomial = {'ja': 1,'nee': 0,np.nan:np.nan}
odi_cleaned['Databases'] = [binomial[item] for item in odi_cleaned['Databases']]
print(odi_cleaned)
```

```python
raw_count = odi_cleaned['Current Program'].value_counts()
#raw_count.plot(kind='bar');
#print(len(raw_count))
odi_cleaner = odi_cleaned
odi_cleaner['Current Program'] = odi_cleaner['Current Program'].str.lower()
#clean_count = odi_cleaner['Current Program'].value_counts()
#clean_count.plot(kind='bar');
#print(len(clean_count))
odi_cleaner = odi_cleaner.replace('&', ' and ',regex=True)
odi_cleaner = odi_cleaner.replace('  and  ', ' and ',regex=True)
banned = ['master','masters','msc','.','@','(joint degree)','(cls)','uva','vu','msc.','in','of']
f = lambda x: ' '.join([item for item in x.split() if item not in banned])
odi_cleaner["Current Program"] = odi_cleaner["Current Program"].apply(f)
odi_cleaner.loc[odi_cleaner['Current Program'].str.contains('artificial intelligence'), 'Current Program'] = 'ai'
odi_cleaner.loc[odi_cleaner['Current Program'].str.contains('computational science'), 'Current Program'] = 'cs'
odi_cleaner.loc[odi_cleaner['Current Program'].str.contains('computer science'), 'Current Program'] = 'cs'
odi_cleaner.loc[odi_cleaner['Current Program'].str.contains('business analytics'), 'Current Program'] = 'ba'
odi_cleaner.loc[odi_cleaner['Current Program'].str.contains('bioinformatics and systems biology'), 'Current Program'] = 'bisb'
odi_cleaner.loc[odi_cleaner['Current Program'].str.contains('bio informatics and systems biology'), 'Current Program'] = 'bisb'
odi_cleaner.loc[odi_cleaner['Current Program'].str.contains('econometrics and operations research'), 'Current Program'] = 'eor'
odi_cleaner.loc[odi_cleaner['Current Program'].str.contains('quantitative risk management'), 'Current Program'] = 'qrm'
print(odi_cleaner['Current Program'].value_counts().index.to_list())
```

```python
print(len(odi_cleaner['Current Program'].value_counts()))
print(len(odi_cleaner))
odi_trimmed = odi_cleaner.dropna()
odi_trimmed = odi_trimmed[odi_trimmed['Current Program'].duplicated(keep=False)]
print(len(odi_trimmed['Current Program'].value_counts()))
print(len(odi_trimmed))
clean_count = odi_trimmed['Current Program'].value_counts()
clean_count.plot(kind='bar');
```

```python
coder = OneHotEncoder()
y = coder.fit_transform(odi_trimmed['Current Program'].to_numpy().reshape(-1,1)).toarray()
frame = pd.DataFrame(y, columns=coder.get_feature_names_out())
try:
    odi_trimmed = odi_trimmed.reset_index(drop=True)
except:
    pass
data = pd.concat([frame, odi_trimmed],axis=1).drop(['Current Program'], axis=1)
odi_without_class = odi_trimmed.drop(['Current Program'], axis=1)
print(odi_without_class)
```

```python
from sklearn import svm
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.feature_selection import VarianceThreshold
from sklearn.feature_selection import SelectKBest
from sklearn.neighbors import KNeighborsClassifier
#clf = KNeighborsClassifier()
cv = KFold(n_splits=10, random_state=1, shuffle=True)
accuracies = []
for i in ['linear','rbf','sigmoid','poly']:
    model = svm.SVC(kernel=i) # Linear Kernel
    scores = cross_val_score(model, odi_without_class, odi_trimmed['Current Program'], scoring='balanced_accuracy', cv=cv, n_jobs=-1)
    print(i,np.mean(scores),np.std(scores))
    accuracies.append([i,np.mean(scores),np.std(scores)])
for i in range(3,51):
    model = KNeighborsClassifier(n_neighbors=i) # Linear Kernel
    scores = cross_val_score(model, odi_without_class, odi_trimmed['Current Program'], scoring='balanced_accuracy', cv=cv, n_jobs=-1)
    print(i,np.mean(scores),np.std(scores))
    accuracies.append([i,np.mean(scores),np.std(scores)])
```
